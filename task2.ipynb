```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('tahirjan.jpg', cv2.IMREAD_GRAYSCALE)
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Grayscale')
plt.subplot(1,2,2); plt.imshow(binary, cmap='gray'); plt.title('Binary')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#example of chain codes
# Load the image and convert to grayscale


img = cv2.imread('tahirjan.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply binary thresholding
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)


# Find contours
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)


import cv2
# Draw the contour on original image
contour_img = img.copy()
cv2.drawContours(contour_img, contours, -1, (0,255,0), 2)

plt.imshow(cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB))
plt.title('Detected Contour')
plt.axis('off')
plt.show()
```

```python
import cv2
img = cv2.imread('tahirjan.jpg')
layer = img.copy()
for i in range(3):
    layer = cv2.pyrDown(layer)
    cv2.imshow(f'Level {i+1}', layer)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
